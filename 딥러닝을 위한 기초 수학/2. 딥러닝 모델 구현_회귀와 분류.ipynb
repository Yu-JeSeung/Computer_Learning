# 딥러닝 모델 구현 - 회귀와 분류

## 1. 딥러닝 모델 구현 - 선형 회귀

```python
# 경고 무시
from warnings import filterwarnings
filterwarnings("ignore")
```

### 1) 환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수 불러오기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2) 데이터 준비

```python
x = np.array([2, 4, 6, 8])     # 공부시간
y = np.array([81, 93, 91, 97]) # 성적
```

### 3) 모델 정의, 컴파일 설정, 모델 학습

```python
model = Sequential()

# 모델 정의 - 입력, 출력 
model.add(Dense(1, input_dim=1, activation='linear'))

# 컴파일 설정 -  옵티마이저 : 경사 하강법(sgd), 오차 함수 : 평균 제곱 오차(mse) 
model.compile(optimizer='sgd', loss='mse')

# 모델 학습
model.fit(x, y, epochs=2000)
```

### 4) 그래프로 확인

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x),'r')    
plt.show()
```

### 5) 모델 테스트

```python
# 7시간 공부, 예상 점수는?
hour = 7
prediction = model.predict([hour])

print("%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다" % (hour, prediction))
```

## 2. 딥러닝 모델 구현 -  다중 선형회귀 모델

### 1) 환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt

#텐서플로의 케라스 API에서 필요한 함수 불러오기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2) 데이터 준비

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]]) # 입력 변수가 2개(학습 시간, 과외 시간)
y = np.array([81, 93, 91, 97]) # 출력 변수 1개(성적)
```

### 3) 모델 정의, 컴파일 설정, 모델 학습

```python
model = Sequential()

# 모델 정의 - 입력 변수가 2개(학습 시간, 과외 시간)이므로 input_dim에 2를 입력합니다. 
model.add(Dense(1, input_dim=2, activation='linear'))

# 모델 컴파일
model.compile(optimizer='sgd' , loss='mse')

# 모델 학습
model.fit(x, y, epochs=2000)
```

### 4) 모델 예측

```python
# 7시간 학습시간, 과외시간 4시간, 예측 점수는?

hour = 7
private_class = 4
prediction = model.predict([[hour, private_class]])

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다" % (hour, private_class, prediction))
```

## 3. 딥러닝 모델 구현 - 로지스틱 회귀 모델(이항 분류)

### 1) 환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수 불러오기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2) 데이터 준비

```python
x = np.array([2, 4, 6, 8, 10, 12, 14]) # 공부 시간
y = np.array([0, 0, 0, 1, 1, 1, 1])    # 합격 여부: 0 - 불합격, 1 - 합격
```

### 3) 모델 정의, 컴파일 설정, 모델 학습

```python
model = Sequential()

# 모델 정의 - 입력, 출력 
model.add(Dense(1, input_dim=1, activation='sigmoid'))

# 컴파일 설정 -  옵티마이저 : 경사 하강법(sgd), 오차 함수 : 교차엔트로피
model.compile(optimizer='sgd' ,loss='binary_crossentropy')

# 모델 학습
model.fit(x, y, epochs=2000)
```

### 4) 그래프로 확인

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x),'r')
plt.show()
```

### 5) 모델 테스트

```python
# 7시간 공부,  합격 예상 확률?
hour = 7
prediction = model.predict([hour])

print("%.f시간을 공부할 경우, 합격 예상 확률은 %.01f%%입니다" % (hour, prediction * 100))
```
